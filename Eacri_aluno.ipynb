### Colunas principais

**pais:** Nome do país parceiro

**sigla_pais:** Código ISO do país (3 letras)

**conti:** Continente

**lingua:** Idioma

**total_proj:** Total de projetos

**primar, sec_terc, social:** Áreas de atuação dos projetos

**VoA:** Indicador composto que mede a liberdade de expressão, participação e prestação de contas.

**GoE:** Medida da qualidade dos serviços públicos e da formulação/implementação de políticas.  

**GNIppp:** Produto Nacional Bruto per capita, ajustado por paridade de poder de compra (PPP)

**gov_soc:** avalia o equilíbrio (ou desequilíbrio) entre projetos de governança e projetos sociais.

**prim_terc:** mostra a predominância relativa de projetos no setor primário versus os setores secundário e terciário.
